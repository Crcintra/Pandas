##### Seleionando Dados dentro de um DataFrame, Adicionando também.

```python
import pandas as pd
reviews = pd.read_csv('/winemag-data-130k-v2.csv', index_col=0)
pd.set_option('display.max_rows', 5)
```

##### Dois tipos de indexação no pandas ("loc e iloc")

```python
#### Iloc é a seleção por posição independente do indice dela, o dataframe vai olhar os dados e vai agir de acordo com  a posição

reviews.iloc[0]
```

```python
### Podemos selecionar através das colunas também, no seguinte formato


reviews.iloc[:,0]
```

```python
### Podemos pegar um pedaço do nosso dataframe também

reviews.iloc[:3,0]
```

```python
reviews.iloc[[0,1,2],0]
```

```python
reviews.iloc[-5:]
```

```python
### Selecionando por registro ou por etiqueta em uma coluna

### Muito útil quando estamos selecionando datas

### Iloc é composição e loc é posição

reviews.loc[0, 'country']
```

```python
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
```

```python
reviews.set_index('title')
```

```python
### Outra forma de fazer seleção de elementos dentro de um dataframe

reviews.country == 'Italy'
```

```python
reviews.loc[reviews.country == 'Italy']
```

```python
reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)]
```

```python
### isin * colocar explicitamente o valor do que vocÊ quer da coluna, se o valor dessa coluna estiver dentro da lista incluir ou não.


reviews.loc[reviews.country.isin(['Italy', 'France'])]
```

```python

### true para valores validos e false para valores nulos.


reviews.loc[reviews.price.notnull()]
```

```python
### Modificar os valores dentro do DataFrame

reviews['critic']  = 'everyone'
reviews['critic']
```

```python
reviews['index_backwards'] =  range(len(reviews), 0, -1)
reviews['index_backwards']
```
